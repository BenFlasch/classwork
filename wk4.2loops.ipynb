```python
# while loop
count = 0
while True: # while this condition is true 
    # run this code block over and over
    if count > 10: 
        break
    print (count)
    count += 1 # count = count + 1 
```

```python
count = 0 # more effiecent version of whats above

while count <= 10:
    print(count)
    count += 1
```

```python
isAgeValid: bool = False


while not isAgeValid:
    age: str = input("enter your age")
    if not age.isnumeric():
        print("only numbers")
        continue # continue to next iteration without reading anything else.
    isAgeValid = True 
    print(f"you chose {age}")

try:
    age = int(age)
except Exception as e:
    print(e)
```

```python
# for loop iterates over a collections
from typing import List 
collection: List[int] = [1,2,3,4,5]

for num in collection: 
    print(num)

# to get index and item, use enumerate
for item in enumerate(collection):
    print(f"{item} is at index {item}")

# deconstruct the variable because enumerate returns (index, seq(index))
for idx, num in enumerate(collection):
    print(f"{num} is at index {idx}")
```

```python
sentence: str = "This is a Python course"
words: List[str] = sentence.split(" ") # going to return a list
used_vowels: List[str] = []
vowels_options: list[str] = ["a", "e", "i", "o", "u", "y"]
for word in words:
    for letter in word:
        if letter in vowels_options:
            used_vowels.append(letter)

print(used_vowels)
```

```python
countries = ["France", "Spain", "USA", "Argentina"]

countries_short = [f"{country [0:3]}." for country in countries]

print(countries_short)
```

```python
# functions take an input (arguments)
# process it, and return an output (return)

def shorten_names(names_list: List[str], chars: int=3) -> List[str]:
    """
    shortens the strings inside a list to chars characters
    
    Args:
        names_list(List[str]): the list of strings you want shortened
        chars (int): the number of characters to shorten each string in the list too
    Return:
        List[str]: a new list of shortened strings
    """
    # cannot read information from outside the function.
    names_list = [f"{name [0:chars]}." for name in names_list]
    return names_list

shorten_names(countries, 2)
shorten_names(countries, chars =2)
```

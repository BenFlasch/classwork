```python
# Strings

# StringIferred capitalization is used for classes
# we do not use stringInferred because of convention
string_inferred = "A super awesome string"
string_declared = str("A declared string")
string_int = '12345'
string_int2 = '567'


print(string_inferred)
print(type(string_inferred))
print(string_int + string_int2) # concatenation
```

```python
# int

int_inferred = 124
int_declared = int(456)
int_coerced = int("789") # coercion from string to int

print(int_inferred, type (int_inferred))
print(int_declared + int_coerced) 
```

```python
# float

float_inferred = 1.0
float_declared = float(2)

print("float_inferred + float_declared =" , float_inferred + float_declared)
print(f"float_inferred: {float_inferred} + float_declared: {float_declared} = {float_inferred + float_declared}")
```

```python
print(f"coerce float_inferred to int: {int(9.999)}")
```

```python
# boolean (true/flase)

bool_inferred = True
bool_declared = bool(0)

print(bool_inferred)
print(bool_declared)
```

```python
upper = string_inferred.upper() # uppercase
lower = string_inferred.lower()
is_numeric = upper.isnumeric()
print(upper)
print(lower)
print(is_numeric)
```

```python
# number math
x = 10
y = 3
print (f"addition = {x + y}") #addition
print (f"subratction = {x - y}") #subtraction
print (f"multiplication = {x * y}") #multiplication
print (f"division = {x // y}") #division
print (f"floor divsion = {x // y}") #floor division
print (f"exponents = {x ** y}") #exponents
print (f"modulus = {x % y}")#modulus
```

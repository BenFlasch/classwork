```python
import pandas as pd
import scipy.stats as stats

url = "https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/penguins_cleaned.csv"
df = pd.read_csv(url)
df.head(11)
```

```python
df.corr(numeric_only = True)
```

```python
r, p = stats.pearsonr(df["body_mass_g"], df["flipper_length_mm"])
r2 = r ** r
print (r,p,r2)
```

```python
r,p = stats.pearsonr(df["flipper_length_mm"], df["bill_length_mm"])
r2 = r** 2
print (r,p,r2)
```
